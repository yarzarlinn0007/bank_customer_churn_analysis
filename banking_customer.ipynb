```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/Bank_Churn.csv')
```

```python
bins = [0, 10000, 30000, float('inf')]
labels = ["Low", "Med", "High"]

df['Income Band'] = pd.cut(df['EstimatedSalary'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind = 'bar')
```

```python
# distribution of unique categorises in categorical columns

categorical_cols = df[["Geography", "Gender", "NumOfProducts", "Income Band"]]

for col in categorical_cols :
  print(f"Value counts for '{col}' :")
  display(df[col].value_counts())
```

```python
# Univariate Analysis

for i , predictor in enumerate(df[["Geography", "Gender", "NumOfProducts", "Income Band"]].columns):
  plt.figure(i)
  sns.countplot(data=df, x=predictor)
```

```python
# Bivariate Analysis

for i , predictor in enumerate(df[["Geography", "Gender", "NumOfProducts", "Income Band"]].columns):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue = 'Gender')
```
